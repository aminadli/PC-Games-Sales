# Game Sales Prediction

```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Extraction

```python
df = pd.read_csv('https://raw.githubusercontent.com/aminadli/PC-Games-Sales/main/Games.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#handling missing value
df.dropna(inplace = True)
```

# Exploratory Data Analysis (EDA)

```python
#analyze number of game genre
df['Genre'].value_counts()
```

**Conclusion:** The highest number of game genre is real-time strategy

```python
#analyze number of game developer
df['Developer'].value_counts()
```

**Conclusion:** The highest number of game developer is Blizzard Entertainment

```python
#analyze number of game publisher
df['Publisher'].value_counts()
```

**Conclusion:** The highest number of game publisher is Electronic Arts

```python
#analyze number of game series
df['Series'].value_counts()
```

**Conclusion:** The highest number of game series is Command & Conquer

```python
print('Maximum of Sales : ', df['Sales'].max())
print('Minimum of Sales : ', df['Sales'].min())
print('Median of Sales :', df['Sales'].median())
print('Average of Sales : ', df['Sales'].mean())
print('Total Sales : ', df['Sales'].sum())
```

```python
#distribution of sales
df['Sales'].hist(color = 'gold')
```

**Conclusion:** The highest distribution of sales between 0 until 5

```python
#analyze of game genre based on average sales
pd.pivot_table(df, index = 'Genre', values = 'Sales', aggfunc = {'Sales' : np.mean}).sort_values('Sales', ascending = False)
```

**Conclusion:** Game genre that has the highest average sales is Sandbox, survival

```python
#analyze of game publisher based on average sales
publisher = pd.pivot_table(df, index = 'Publisher', values = 'Sales', aggfunc = {'Sales' : np.mean}).sort_values(
    'Sales', ascending = False)
publisher.head()
```

**Conclusion:** Game publisher that has the highest average sales is Mojang Studios

```python
#analyze of game developer based on average sales
developer = pd.pivot_table(df, index = 'Developer', values = 'Sales', aggfunc = {'Sales' : np.mean}).sort_values(
    'Sales', ascending = False)
developer.head()
```

**Conclusion:** Game developer that has the highest average sales is Mojang Studios

```python
#analyze of game series based on average sales
series = pd.pivot_table(df, index = 'Series', values = 'Sales', aggfunc = {'Sales' : np.mean}).sort_values(
    'Sales', ascending = False)
series.head()
```

**Conclusion:** Game series that has the highest average sales is Minecraft

```python
#visualize game genre based on average sales
plt.figure(figsize = (13,5))
df.groupby('Genre')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'darkorange')
plt.title('Game Genre based on Average Sales', fontsize = 15, color = 'b', pad = 12)
plt.xlabel('Genre', fontsize = 12)
plt.ylabel('Average Sales', fontsize = 12)
plt.show()
```

```python
#visualize game publisher based on average sales
plt.figure(figsize = (15,5))
df.groupby('Publisher')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'lime')
plt.title('Game Publisher based on Average Sales', fontsize = 15, color = 'b', pad = 12)
plt.xlabel('Publisher', fontsize = 12)
plt.ylabel('Average Sales', fontsize = 12)
plt.show()
```

```python
#visualize game developer based on average sales
plt.figure(figsize = (15,5))
df.groupby('Developer')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'orangered')
plt.title('Game Developer based on Average Sales', fontsize = 15, color = 'b', pad = 12)
plt.xlabel('Developer', fontsize = 12)
plt.ylabel('Average Sales', fontsize = 12)
plt.show()
```

```python
#visualize game series based on average sales
plt.figure(figsize = (18,5))
df.groupby('Series')['Sales'].mean().sort_values(ascending = False).plot(kind = 'bar', color = 'blueviolet')
plt.title('Game Series based on Average Sales', fontsize = 15, color = 'r', pad = 12)
plt.xlabel('Series', fontsize = 12)
plt.ylabel('Average Sales', fontsize = 12)
plt.show()
```

```python
#analyze game sales based on release
sales = pd.pivot_table(df, index = 'Release', values = 'Sales', aggfunc = {'Sales' : np.mean}).sort_values('Sales',
                                                                                                           ascending = False)
sales.head()
```

```python
#visualize game sales trend based on average sales
chart = sales.sort_values(by = 'Sales', ascending = False).plot(kind = 'line', color = 'coral')
chart.set_xticklabels(chart.get_xticklabels(), rotation = 45)
plt.title('Game Sales Trend based on Game Release', fontsize = 15, color = 'b', loc = 'center', pad = 12)
plt.xlabel('Release', fontsize = 12)
plt.ylabel('Average Sales', fontsize = 12)
plt.show()
```
